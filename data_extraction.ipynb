We get the list of the files we want to read to extract the articles

```python
import os
files = []
for file in os.listdir("./data") :
    if file.endswith(".TXT") :
        files.append(file)
files
```

Now, in each document we have around 500 articles. We want to extract every articles and save it in a file. we will have one file with the content of the article and an other file related to it with the name of the newspaper and the date.

```python
#index doc
numDoc = 1

for i in files :
    #Open TXT file and read it
    txtfile = open("./data/"+i)
    content = txtfile.read()
    
    content_line = content.split('\n')
    i = 0
    while i < len(content_line) :
        
        article_content = open("./data/content/article_"+str(numDoc)+'.txt' , 'w')
        article_ref = open("./data/ref/article_"+str(numDoc)+'.txt' , 'w')
        end = False
        
        if i+4 >= len(content_line) : 
            break 
        nameNewspaper = content_line[i+4]
        newspaper = nameNewspaper.split(' ')
        print(newspaper)
        for n in newspaper :
            if n != '' :
                article_ref.write(n+' ')
        print(nameNewspaper)
        article_ref.write('\n')
        date = content_line[i+6]
        date1 = date.split(' ')
        for d in date1 :
            if d != '' :
                article_ref.write(d+' ')
        print(date)
        article_ref.close()
        
        i+=7
        print('-------------')

        while content_line[i] != '' or content_line[i+1] != '' :
            print(content_line[i])
            i+=1
        print("===========")
        i+=2
        
        while end == False : 
            line = content_line[i].split(' ') 
            #print(i)
            if line[0] != "LOAD-DATE:" : 
                article_content.write(content_line[i])
                i+=1
            else : 
                print("L")
                end = True
                i+=1
                break
        end = False 
        while end == False :
            line = content_line[i].split(' ')
            #print(line)
            for k in line : 
                if k == "Copyright" :
                    print("here")
                    i+=1
                    end = True
                    break 
            if end == False :
                i+= 1
        #i+=1
        print("end")
        while content_line[i] != '' :
            print(i)
            i+=1
        i+=1
        print("THE end of doc "+ str(numDoc))
        article_content.close()
        numDoc+=1
    
    txtfile.close()
```
